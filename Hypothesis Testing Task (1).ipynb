Hypothesis Testing

Is there a statistically significant difference between the mean age of patients who lived and did not survive after the surgery?

```python
df.groupby("Patient_Status").agg({"Age": "mean"})
```

- H0: there is no statistically significant difference between living and non-living.
- H1: there is significant difference.

```python
ttest, pval = mannwhitneyu(df[df["Patient_Status"] == 'Alive']["Age"], df[df["Patient_Status"] == 'Dead']["Age"])
print(pval)
```

```python
if pval < 0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

Another Hypothesis Testing:

- Null Hypothesis (H0): There is no significant relationship or difference between age and protein level 1.
- Alternative Hypothesis (H1): There is a significant relationship or difference between age and protein level 1.

```python
df.groupby("Age").agg({"Protein1": "mean"})
```

```python
young_age_group = df[df['Age'] < 30]['Protein1']
old_age_group = df[df['Age'] >= 30]['Protein1']
```

```python
ttest,pval = stats.ttest_ind(young_age_group, old_age_group)
print(pval)
```

```python
if pval < 0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
