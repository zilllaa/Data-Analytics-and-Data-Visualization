creating two line plots using Matplotlib. It also sets labels for the x-axis and y-axis, adds a title to the plot, and includes a legend to distinguish between the two datasets.

```python
plt.plot(years, apples)
plt.plot(years, tangerines)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Tangerines'])
```

creating two line plots using Matplotlib with marker points. It also sets labels for the x-axis and y-axis, adds a title to the plot, and includes a legend to distinguish between the two datasets.

```python
plt.plot(years, apples, marker='o')
plt.plot(years, tangerines, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Tangerines'])
```

```python
#creating a figure with a specific size.

plt.figure(figsize=(12,6)) #sets up figure with a size of 12 inches in width and 6 inches in height

plt.plot(years, tangerines, marker='o')
plt.title("Yield od Tangerines (tons per hectare)")
```

#### Changing the size of bins using numpy

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
penguins.head(3)
```

```python
import numpy as np

# specify the number of bins
plt.hist(penguins.flipper_length_mm, bins=np.arange(170, 200, 0.5))
```

Bins is a parameter specifying the bin edges. In this case, we're creating bins with edges ranging from 170 to 200 (exclusive) with bin widths of 0.5.

```python
#bins of unequal size
plt.hist(penguins.flipper_length_mm, bins=[160, 190, 200, 250])
```

In this example, there are three intervals used to group the flipper length data. All data falling within interval 1 (170-200) are grouped together, as are those in interval 2 (200-230) and interval 3 (230-260).

It's possible to create several histograms on a single chart, similar to how we use line charts. To avoid one histogram obscuring the others, we can decrease the opacity of each histogram. First, we generate distinct histograms for each penguin species.

```python
penguins['species'].unique()
```

```python
adelie = penguins[penguins.species == 'Adelie']
chinstrap = penguins[penguins.species == 'Chinstrap']
gentoo = penguins[penguins.species == 'Gentoo']
```

```python
plt.hist(adelie.flipper_length_mm, alpha = 0.4)
plt.hist(chinstrap.flipper_length_mm, alpha = 0.4)
plt.hist(gentoo.flipper_length_mm, alpha = 0.4)
```

### Seaborn
Seaborn is a popular data visualization library built on top of Matplotlib in Python. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn simplifies the process of creating complex visualizations, such as statistical plots, heatmaps, pair plots, and more, with less code.Seaborn complements Matplotlib and is often used for data analysis and exploration due to its simplified and informative plotting capabilities.

```python
sns.set_style("whitegrid")
```

The sns.set_style("whitegrid") command in Seaborn is used to set the plotting style for your Seaborn visualizations. In this specific case, it sets the style to "whitegrid," which is one of the built-in styles provided by Seaborn. The "whitegrid" style includes a white background with grid lines for both the x and y axes. This style is often used for creating clean and easily readable visualizations.

#### Histogram
- The histogram visualization is also bar-shaped.
- However, a histogram is a type of graph that explains frequencies based on two numerical data.

```python
iris_df=sns.load_dataset('iris')
iris_df.head()
```

```python
iris_df.sepal_width
```

```python
plt.title('Distribution of Sepal Width')
plt.hist(iris_df.sepal_width)
```

bin sometimes called a class interval is a way of sorting data in a histogram.

```python
#specifing the number of bins
plt.hist(iris_df.sepal_width, bins=10)
```

##### Pro Tips
- Histograms are commonly used to view the frequency of continuous numerical data.
- Bar charts are commonly used to view the frequency of categorical data.
